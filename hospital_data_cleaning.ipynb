```python
import pandas as pd
import numpy as np
import os
```

```python
import kagglehub

path = kagglehub.dataset_download("nudratabbas/hospital-records-for-data-cleaning-medium")
path
```

```python
import os
os.listdir(path)
```

```python
df = pd.read_csv(os.path.join(path, "hospital_patients_real_world.csv"))
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df[df[['Age', 'Gender', 'Diagnosis']].isna().any(axis=1)].shape
```

```python
df[['Age', 'Gender', 'Diagnosis']].isna().sum(axis=1).value_counts().sort_index()
```

```python
df = df[df[['Age', 'Gender', 'Diagnosis']].isna().sum(axis=1) <= 1]
df.shape
```

```python
df[['Age', 'Gender', 'Diagnosis']].isna().sum()
```

```python
age_median = df['Age'].median()
df['Age'] = df['Age'].fillna(age_median)
```

```python
df['Age'].isna().sum()
```

```python
df['Gender'] = df['Gender'].fillna('Unknown')
```

```python
df['Gender'].isna().sum()
```

```python
df['Diagnosis'] = df['Diagnosis'].fillna('Unknown')
```

```python
df['Diagnosis'].isna().sum()
```

```python
df.to_csv("hospital_patients_clean.csv", index=False)
```
